# **Missing Values**

```python
#check for all null values
df.isnull().sum()
```

```python
(df.isnull().sum() / len(df)) * 100
```

The dataset contains missing values in the following columns:
state, work_interfere, and comments.

The **state** column contains missing values mainly for respondents who are not from the United States.
Instead of removing these records, a placeholder value is used to preserve important survey responses.

```python
# Fill 'state' missing values
df['state'] = df['state'].fillna('Not Specified')
```

**Insight:**

This approach ensures that geographic information remains consistent and prevents unnecessary data loss while clearly indicating that state-level data is unavailable for some respondents.

The **work_interfere** column is categorical and represents how mental health affects work performance. Filling missing values with the most frequent response (mode) maintains the overall distribution of the data.

```python
# Fill 'work_interfere' with mode
df['work_interfere'] = df['work_interfere'].fillna(df['work_interfere'].mode()[0])
```

**Insight:**

This helps retain records for analysis and avoids bias that could occur if rows with missing values were removed, while still reflecting the most common employee experience.

The **comments** column contains a large number of missing values and unstructured free-text data, making it less suitable for quantitative analysis.

```python
# Drop 'comments' column (too many missing values)
df.drop(columns=['comments'], inplace=True)
```

**Insight:**

Removing this column simplifies the dataset and allows the analysis to focus on structured variables that provide clearer and more reliable insights.

```python
df.columns
```

**Clean Age Column**

Removed unrealistic age values. Extreme or incorrect age values can distort visualizations and statistical analysis.

```python
#Clean Age Column
df = df[(df['Age'] >= 18) & (df['Age'] <= 100)]
```

**Insight:**

Ensures age-based analysis represents realistic working professionals.

**Clean and Standardize Gender**

Removed unrealistic age values.Outliers can skew analysis and visualizations.Outliers can skew analysis and visualizations.

```python
#Clean and Standardize Gender
df['Gender'] = df['Gender'].str.lower()

df['Gender'] = df['Gender'].replace({
    'male': 'Male', 'm': 'Male', 'man': 'Male',
    'female': 'Female', 'f': 'Female', 'woman': 'Female'
})

df['Gender'] = df['Gender'].apply(
    lambda x: 'Other' if x not in ['Male', 'Female'] else x
)
```

**Insight:**

Ensures age-related analysis reflects realistic working professionals.

**Checking for Duplicates in dataset**

```python
#Checking for Duplicates
df.duplicated().sum()
```

When we checked for duplicate rows, the result was zero, which means the dataset does not contain any duplicate records.

 **Standardizing Gender Values**

 Merged multiple gender representations into common categories.Inconsistent gender entries affect grouping, counting, and visualization accuracy.

```python
#Standardizing Gender Values
bins = [18, 25, 35, 45, 60, 100]
labels = ['18-25', '26-35', '36-45', '46-60', '60+']

df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

**Insight:**

Standardization allows meaningful gender-based comparisons and highlights participation imbalance.

**Country-wise Mental Health Treatment**

Grouped the dataset by Country and analyzed the distribution of employees who have sought mental health treatment. To analyzing treatment patterns by country helps identify regional differences in mental health behavior.

```python
country_treatment = df.groupby('Country')['treatment'].value_counts(normalize=True).unstack()

country_treatment.head()
```

**Insight:**

Employees from developed countries report higher rates of mental health treatment, likely due to better awareness and access to healthcare. In contrast, lower reported treatment rates in some regions may reflect social stigma or limited mental health resources rather than lower prevalence.

As we can see dataset is clean and ready for further analysis, the manipulation performed on given data are as follows:

**Handling Missing Values**

*   Null values in the state column were replaced with "Not Specified" to preserve non-US responses.
*   Missing values in the work_interfere column were filled using the mode to maintain categorical distribution.
*   The comments column was removed due to a high number of missing values and unstructured text data.


**Data Cleaning**
*   Unrealistic values in the Age column were removed to ensure meaningful analysis.
*   Inconsistent entries in the Gender column were standardized into common categories.


**Handling Duplicates**


*   Duplicate records were checked and removed to avoid biased or repeated survey responses.

# **14.   Correlation Heatmap**

*   A correlation heatmap is used to measure the strength and direction of relationships between numerical variables.
*   It helps identify positive, negative, or no correlation at a glance.
*   Useful for detecting redundant features, understanding dependencies, and supporting feature selection.


Selected numerical columns from the dataset and visualized their correlations using a heatmap.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

**Insight:**

*   Most numerical variables show weak correlation, indicating that they capture different aspects of mental health.
*  Age shows little to no correlation with mental health treatment, suggesting mental health issues affect all age groups.
*   The absence of strong correlations highlights the importance of categorical variables in this dataset.

**Business Impact**

Shows that mental health is influenced by multiple factors, supporting holistic policy decisions instead of single-factor solutions.



# **15.   Pair Plot**

A pair plot helps visualize the distribution of numerical variables and their relationships with other variables at the same time. It is useful to check whether a numerical feature like Age has any visible pattern or influence on mental health outcomes.




```python
# Pair plot using Age and treatment
sns.pairplot(
    df,
    vars=['Age'],
    hue='treatment',
    diag_kind='hist',
    height=4
)

plt.suptitle("Pair Plot of Age vs Mental Health Treatment", y=1.02)
plt.show()
```

**Insights:**

The chart shows that mental health treatment is spread across all age groups, with no strong age-specific pattern. This indicates that mental health challenges are not limited to a particular age range.

**Business Impact**

Since mental health issues affect employees across all ages, organizations should design inclusive mental health policies and support systems rather than targeting only specific age groups. This leads to better employee well-being and improved productivity.
